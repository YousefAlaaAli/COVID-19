## 6. VISUALIZING WORLD DATA

### Confirmed cases (Total Cases) all over the world

```python
df_world_over_time = df_world.groupby(["date"])[["total_cases","new_cases","total_deaths","new_deaths"]].sum()
df_world_over_time
```

```python
df_world_over_time = df_world.groupby(["date"])[["total_cases","new_cases","total_deaths","new_deaths"]].sum().reset_index()
df_world_over_time
```

```python
# Using line-plot

plt.figure(figsize=(20,10))  

#plt.plot(df_world_over_time['date'], df_world_over_time['total_cases'])
plt.plot(df_world_over_time.index, df_world_over_time['total_cases'])

plt.title('Evolution of Confirmed Covid-19 cases over time in the word', fontsize=16)
plt.xlabel('No. of days', fontsize=16)
plt.ylabel('Confirmed cases', fontsize=1)
plt.show()
```

```python
# Using line-plot
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time['date'], df_world_over_time['total_deaths'])

#plt.plot(df_world_over_time.index, df_world_over_time['total_deaths'])

plt.title('Evolution of Confirmed Covid-19 cases over time in the word', fontsize=16)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Confirmed cases', fontsize=16)
plt.show()
```

### Total deaths cases evolution over time

```python
# Using line-plot
plt.figure(figsize=(20,10))

#plt.plot(df_world_over_time.index, df_world_over_time['total_deaths'])
plt.plot(df_world_over_time['date'], df_world_over_time['total_cases'])

plt.title('Evolution of Covid-19 Deaths cases over time in the world', fontsize=16)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Number of Deaths', fontsize=16)
```

### New cases all over the world

```python
# Using line-plot
plt.figure(figsize=(12,5))
#plt.bar(df_world_over_time.index, df_world_over_time['new_cases'])
plt.plot(df_world_over_time['date'], df_world_over_time['total_cases'])

plt.title('Evolution of Covid-19 New Cases over time in the world', fontsize=16)
plt.xlabel('Year', fontsize=16)
plt.ylabel('New Cases', fontsize=16)
```

### Calculate Total Recovered

```python
# Calculate total_recovered
df_world_over_time['total_recovered'] = df_world_over_time['total_cases'] - df_world_over_time['total_deaths']

# Create the line plot
plt.figure(figsize=(20, 10))
plt.plot(df_world_over_time['date'], df_world_over_time['total_recovered'], label='Total Recovered', color='green')

# Add title and labels
plt.title('Evolution of Covid-19 Recovered Cases Over Time in the World', fontsize=16)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Number of Recovered Cases', fontsize=16)
plt.legend()  # Show the legend
plt.grid(True)  # Enable grid for better visualization

# Display the plot
plt.show()
```

### Putting it all together

```python
# Using line-plot
plt.figure(figsize=(20,10))
#plt.plot(df_world_over_time.index, df_world_over_time['total_cases'], label='Confirmed')
plt.plot(df_world_over_time.index, df_world_over_time['total_deaths'], label='Total Deaths')
plt.plot(df_world_over_time.index, df_world_over_time['new_cases'], label='New')
plt.plot(df_world_over_time.index, df_world_over_time['new_deaths'], label='New Deaths')
plt.legend(loc=0)
plt.show()
```
